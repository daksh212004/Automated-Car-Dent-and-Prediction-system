```python
import xml.etree.ElementTree as ET
from pathlib import Path
import os

# Discover all unique classes in XML files
def get_all_classes(annot_path):
    classes = set()
    xml_files = [f for f in os.listdir(annot_path) if f.endswith('.xml')]
    for xml_file in xml_files:
        tree = ET.parse(annot_path / xml_file)
        for obj in tree.findall('object'):
            classes.add(obj.find('name').text)
    return sorted(list(classes))

# Define class names (all classes from dataset)
local_annot_path = Path('/content/dataset/annot_chem4')
class_names = get_all_classes(local_annot_path)
print(f"Detected classes: {class_names}")

# Optional: Filter to specific classes if desired
# class_names = ['Scratch_or_spot', 'Dent']
# print(f"Using filtered classes: {class_names}")

class_to_id = {name: idx for idx, name in enumerate(class_names)}

def convert_voc_to_yolo(xml_file, img_path, output_path):
    tree = ET.parse(xml_file)
    root = tree.getroot()

    # Get image dimensions
    size = root.find('size')
    width = int(size.find('width').text)
    height = int(size.find('height').text)

    # Get image filename and normalize extension
    img_file = root.find('filename').text
    img_file_jpg = img_file if img_file.endswith('.jpg') else img_file.replace('.jpeg', '.jpg').replace('.png', '.jpg')

    # Check if image exists
    if not os.path.exists(img_path / img_file_jpg):
        print(f"Image not found for {xml_file}: {img_file_jpg}")
        return False

    # Parse objects
    yolo_lines = []
    for obj in root.findall('object'):
        name = obj.find('name').text
        if name not in class_to_id:
            continue  # Skip unknown classes
        class_id = class_to_id[name]

        # Get bounding box
        bbox = obj.find('bndbox')
        xmin = int(bbox.find('xmin').text)
        ymin = int(bbox.find('ymin').text)
        xmax = int(bbox.find('xmax').text)
        ymax = int(bbox.find('ymax').text)

        # Convert to YOLO format
        x_center = (xmin + xmax) / 2 / width
        y_center = (ymin + ymax) / 2 / height
        box_width = (xmax - xmin) / width
        box_height = (ymax - ymin) / height

        yolo_lines.append(f"{class_id} {x_center:.6f} {y_center:.6f} {box_width:.6f} {box_height:.6f}")

    # Write YOLO annotation file (even if empty)
    output_file = output_path / f"{xml_file.stem}.txt"
    with open(output_file, 'w') as f:
        if yolo_lines:
            f.write('\n'.join(yolo_lines))
            return True
        else:
            f.write('')  # Create empty file
            return False

# Convert all annotations
local_img_path = Path('/content/dataset/images_chem4')
local_yolo_annot_path = Path('/content/dataset/yolo_labels')
xml_files = sorted([f for f in os.listdir(local_annot_path) if f.endswith('.xml')])
valid_conversions = 0
for xml_file in xml_files:
    if convert_voc_to_yolo(local_annot_path / xml_file, local_img_path, local_yolo_annot_path):
        valid_conversions += 1

print(f"Converted {len(xml_files)} XML files to YOLO format ({valid_conversions} with valid annotations)")
```

```python
import os
from pathlib import Path
import random

# Paths
local_img_path = Path('/content/dataset/images_chem4')
local_yolo_annot_path = Path('/content/dataset/yolo_labels')
local_base_path = Path('/content/dataset')

# Create train and val directories
train_img_path = local_base_path / 'train/images'
train_label_path = local_base_path / 'train/labels'
val_img_path = local_base_path / 'val/images'
val_label_path = local_base_path / 'val/labels'
os.makedirs(train_img_path, exist_ok=True)
os.makedirs(train_label_path, exist_ok=True)
os.makedirs(val_img_path, exist_ok=True)
os.makedirs(val_label_path, exist_ok=True)

# Get images with annotations
img_files = sorted([f for f in os.listdir(local_img_path) if f.endswith('.jpg')])
annot_files = sorted([f for f in os.listdir(local_yolo_annot_path) if f.endswith('.txt')])
valid_imgs = [img for img in img_files if img.replace('.jpg', '.txt') in annot_files]

# Split dataset (80% train, 20% val)
random.seed(42)
random.shuffle(valid_imgs)
split_idx = int(0.8 * len(valid_imgs))
train_imgs = valid_imgs[:split_idx]
val_imgs = valid_imgs[split_idx:]

# Copy files to train/val directories
for img in train_imgs:
    !cp {local_img_path / img} {train_img_path}
    annot = img.replace('.jpg', '.txt')
    !cp {local_yolo_annot_path / annot} {train_label_path}
for img in val_imgs:
    !cp {local_img_path / img} {val_img_path}
    annot = img.replace('.jpg', '.txt')
    !cp {local_yolo_annot_path / annot} {val_label_path}

# Create data.yaml
data_yaml = f"""
train: {train_img_path}
val: {val_img_path}
nc: {len(class_names)}
names: {class_names}
"""

with open(local_base_path / 'data.yaml', 'w') as f:
    f.write(data_yaml)

print("Created data.yaml with train/val split")
print(f"Train images: {len(train_imgs)}, Val images: {len(val_imgs)}")
```

```python
def get_bbox(xml_file):
    tree = ET.parse(xml_file)
    img_file = tree.find('filename').text
    damages = [obj.find('name').text for obj in tree.findall('object')]
    bboxes = [[int(obj.find('bndbox').find(tag).text) for tag in ['xmin', 'ymin', 'xmax', 'ymax']]
              for obj in tree.findall('object')]
    return img_file, damages, bboxes

def draw_bbox(img_file, damages, bboxes):
    img = cv2.imread(str(local_img_path / img_file))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    fig = plt.figure(figsize=(10, 10))

    print(f"Image File: {img_file}")
    print(f"Image Dimensions: {img.shape}")
    print("Damage Type --> Bounding Box:")
    for d, bb in zip(damages, bboxes):
        print(f"\t{d} --> {bb}")
        img = cv2.rectangle(img, (bb[0], bb[1]), (bb[2], bb[3]), (255, 0, 255), 3)
        img = cv2.putText(img, d, (bb[0], bb[1]-5), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 0), 1, cv2.LINE_AA)

    plt.imshow(img)
    plt.axis('off')
    plt.show()

# Display one sample image
sample_xml = xml_files[0]  # First XML file
img_file, damages, bboxes = get_bbox(local_annot_path / sample_xml)
draw_bbox(img_file, damages, bboxes)
```

```python
def test_model(img_path, model, threshold=0.3):
    results = model.predict(img_path, conf=threshold)
    img = cv2.imread(str(img_path))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    print(f"Testing Image: {img_path.name}")
    fig = plt.figure(figsize=(10, 10))

    for result in results:
        boxes = result.boxes.xyxy.cpu().numpy()  # Bounding boxes
        scores = result.boxes.conf.cpu().numpy()  # Confidence scores
        classes = result.boxes.cls.cpu().numpy().astype(int)  # Class IDs

        for box, score, cls in zip(boxes, scores, classes):
            xmin, ymin, xmax, ymax = map(int, box)
            label = f"{class_names[cls]} ({score:.2f})"
            img = cv2.rectangle(img, (xmin, ymin), (xmax, ymax), (255, 0, 255), 3)
            img = cv2.putText(img, label, (xmin, ymin-5), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 0), 1, cv2.LINE_AA)

    plt.imshow(img)
    plt.axis('off')
    plt.show()

# Test on a sample image
test_img = local_img_path / 'akhand_b43_157.jpg'  # Adjust if needed
test_model(test_img, model, threshold=0.3)
```

```python
import cv2
import matplotlib.pyplot as plt
from pathlib import Path
from ultralytics import YOLO
import yaml

# Load the trained model
model = YOLO('/content/runs/ skorpio_yolov8/weights/best.pt')

def test_model(img_path, model, class_names, threshold=0.3):
    # Run inference
    results = model.predict(img_path, conf=threshold, verbose=False)
    img = cv2.imread(str(img_path))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    print(f"Testing Image: {img_path.name}")
    fig = plt.figure(figsize=(10, 10))

    # Draw bounding boxes
    for result in results:
        boxes = result.boxes.xyxy.cpu().numpy()
        scores = result.boxes.conf.cpu().numpy()
        classes = result.boxes.cls.cpu().numpy().astype(int)

        for box, score, cls in zip(boxes, scores, classes):
            xmin, ymin, xmax, ymax = map(int, box)
            label = f"{class_names[cls]} ({score:.2f})"
            img = cv2.rectangle(img, (xmin, ymin), (xmax, ymax), (255, 0, 255), 3)
            img = cv2.putText(img, label, (xmin, ymin-10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 0), 2, cv2.LINE_AA)

    plt.imshow(img)
    plt.axis('off')

    # Save the annotated image
    output_path = Path('/content/runs/skorpio_yolov8/test_samples') / img_path.name
    output_path.parent.mkdir(parents=True, exist_ok=True)
    plt.savefig(output_path, bbox_inches='tight', pad_inches=0)
    plt.close()
    print(f"Saved annotated image to: {output_path}")

# Test on a sample validation image
val_img_path = Path('/content/dataset/val/images')
sample_img = list(val_img_path.glob('*.jpg'))[0] if list(val_img_path.glob('*.jpg')) else None
if sample_img:
    with open('/content/dataset/data.yaml', 'r') as f:
        data_yaml = yaml.safe_load(f)
    class_names = data_yaml['names']
    test_model(sample_img, model, class_names, threshold=0.3)
else:
    print("No validation images found in /content/dataset/val/images")
```

```python
import torch
from ultralytics import YOLO
from pathlib import Path

# Load the trained model
model = YOLO('/content/runs/ skorpio_yolov8/weights/best.pt')

# Run validation
results = model.val(
    data='/content/dataset/data.yaml',
    imgsz=640,
    batch=16,
    device=0 if torch.cuda.is_available() else 'cpu',
    project='/content/runs',
    name='skorpio_yolov8_val'
)

# Extract and print metrics
metrics = results.results_dict
print("\nValidation Metrics:")
print(f"mAP@0.5: {metrics['metrics/mAP50(B)']:.4f}")
print(f"mAP@0.5:0.95: {metrics['metrics/mAP50-95(B)']:.4f}")
print(f"Precision: {metrics['metrics/precision(B)']:.4f}")
print(f"Recall: {metrics['metrics/recall(B)']:.4f}")

# Save metrics to a file
output_path = Path('/content/runs/skorpio_yolov8_val/metrics.txt')
output_path.parent.mkdir(parents=True, exist_ok=True)
with open(output_path, 'w') as f:
    for key, value in metrics.items():
        f.write(f"{key}: {value:.4f}\n")
print(f"Saved metrics to: {output_path}")
```

```python
import matplotlib.pyplot as plt
import cv2
from pathlib import Path

def display_image(img_path, title):
    img = cv2.imread(str(img_path))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(8, 8))
    plt.imshow(img)
    plt.title(title)
    plt.axis('off')
    plt.show()

# Paths to validation outputs
val_results_path = Path('/content/runs/skorpio_yolov8_val')

# Display confusion matrix
confusion_matrix_path = val_results_path / 'confusion_matrix.png'
if confusion_matrix_path.exists():
    display_image(confusion_matrix_path, "Confusion Matrix")
else:
    print("Confusion matrix not found")

# Display precision-recall curve
pr_curve_path = val_results_path / 'PR_curve.png'
if pr_curve_path.exists():
    display_image(pr_curve_path, "Precision-Recall Curve")
else:
    print("PR curve not found")
```
